```python
import os
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Define a custom NLTK data directory
nltk_data_dir = os.path.expanduser('~/nltk_data')

# Add the custom directory to the NLTK data path
nltk.data.path.append(nltk_data_dir)

# Function to download NLTK resources with error handling
def download_nltk_resources():
    try:
        nltk.download('stopwords', download_dir=nltk_data_dir)
        nltk.download('punkt', download_dir=nltk_data_dir)
    except LookupError as e:
        print(f"Error downloading NLTK resources: {e}")

# Download the necessary resources
download_nltk_resources()

# Check if resources are available
if not os.path.exists(os.path.join(nltk_data_dir, 'corpora/stopwords.zip')):
    nltk.download('stopwords', download_dir=nltk_data_dir)

if not os.path.exists(os.path.join(nltk_data_dir, 'tokenizers/punkt.zip')):
    nltk.download('punkt', download_dir=nltk_data_dir)

# Define the set of stop words
stop_words = set(stopwords.words('english'))

# Function to clean tweets
def clean_tweet(tweet):
    if isinstance(tweet, str):
        tweet = tweet.lower()
        tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
        tweet = re.sub(r'\@\w+|\#', '', tweet)
        tweet = re.sub(r'[^A-Za-z0-9\s]+', '', tweet)
        tweet_tokens = word_tokenize(tweet)
        filtered_words = [word for word in tweet_tokens if word not in stop_words]
        return " ".join(filtered_words)
    else:
        return ""

# Assuming 'df' is your DataFrame and it has a 'text' column
df['cleaned_tweet'] = df['text'].apply(clean_tweet)
```

Handling Missing Value 

```python
df.isnull().sum()
df.dropna(subset=['cleaned_tweet'],inplace=True)
```

Text Normalization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize_tweet(tweet):
    return " ".join([lemmatizer.lemmatize(word) for word in tweet.split()])
df['normalized_tweet'] = df['cleaned_tweet'].apply(lemmatize_tweet)   
```

Save Cleaned data

```python
df.to_csv('cleaned_metoo_tweets_dec2017.csv', index=False)
```
